```
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_validate
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import matthews_corrcoef
from sklearn.metrics import make_scorer
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import make_pipeline
```

```
path_dataset = '/content/drive/My Drive/UFPR/aplicacao_ciencia_dados/trabalho_disciplina/dataset/training_dataset.csv'
data = pd.read_csv(path_dataset)
```

```
X = data[data.columns[:-1]]
y = data[data.columns[-1:]]
```

```
X_1 = X[['age_of_domain', 'DNSRecord', 'web_traffic', 'Page_Rank','Google_Index', 'Links_pointing_to_page', 'Statistical_report']]
X_2 = X[['Redirect', 'on_mouseover', 'RightClick', 'popUpWidnow', 'Iframe']]
X_3 = X[['Request_URL', 'URL_of_Anchor', 'Links_in_tags', 'SFH', 'Submitting_to_email', 'Abnormal_URL']]
X_4 = X[['having_IP_Address', 'URL_Length', 'Shortining_Service', 'having_At_Symbol', 'double_slash_redirecting', 'Prefix_Suffix', 'having_Sub_Domain', 'SSLfinal_State', 'Domain_registeration_length', 'Favicon', 'port', 'HTTPS_token',]]
```

```
soma_1 = X_1.sum(axis=1).copy()
soma_2 = X_2.sum(axis=1).copy()
soma_3 = X_3.sum(axis=1).copy()
soma_4 = X_4.sum(axis=1).copy()
```

```
X_soma = pd.DataFrame({'1':soma_1, '2':soma_2, '3':soma_3, '4':soma_4})
```

```
X_train, X_test, y_train, y_test = train_test_split(X_soma, y, test_size=0.2, random_state=42)
```

```
skf = StratifiedKFold(n_splits=5)
```

#RandomForest

```
model_random =  make_pipeline(MinMaxScaler(feature_range=(-1,1)), RandomForestClassifier())
```

```
result_random = cross_validate(model_random, X_train, y_train, scoring={'f1_score' : make_scorer(f1_score),'accuracy_score': make_scorer(accuracy_score),'matthews_corrcoef':make_scorer(matthews_corrcoef)}, cv=skf, n_jobs=-1)
```

```
pd_result_random=pd.DataFrame(result_random)
```

```
pd_result_random
```

```
pd_result_random.mean()
```

#Logistic

```
model_logistic = make_pipeline(MinMaxScaler(feature_range=(-1,1)), LogisticRegression())
```

```
result_logistic = cross_validate(model_logistic, X_train, y_train, scoring={'f1_score' : make_scorer(f1_score),'accuracy_score': make_scorer(accuracy_score),'matthews_corrcoef':make_scorer(matthews_corrcoef)}, cv=skf, n_jobs=-1)
```

```
pd_result_logistic = pd.DataFrame(result_logistic)
```

```
pd_result_logistic
```

```
pd_result_logistic.mean()
```

#Linear SVM

```
model_svm =  make_pipeline(MinMaxScaler(feature_range=(-1,1)), LinearSVC())
```

```
result_svm = cross_validate(model_svm, X_train, y_train, scoring={'f1_score' : make_scorer(f1_score),'accuracy_score': make_scorer(accuracy_score),'matthews_corrcoef':make_scorer(matthews_corrcoef)}, cv=skf, n_jobs=-1)
```

```
pd_result_svm = pd.DataFrame(result_svm)
```
