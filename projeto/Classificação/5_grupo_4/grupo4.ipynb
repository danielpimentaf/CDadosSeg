```
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_validate
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import matthews_corrcoef
from sklearn.metrics import make_scorer
```

```
path_dataset = '/content/drive/My Drive/UFPR/aplicacao_ciencia_dados/trabalho_disciplina/dataset/training_dataset.csv'
data = pd.read_csv(path_dataset)
```

```
X = data[data.columns[:-1]]
y = data[data.columns[-1:]]
```

```
X_4 = X[['having_IP_Address', 'URL_Length', 'Shortining_Service', 'having_At_Symbol', 'double_slash_redirecting', 'Prefix_Suffix', 'having_Sub_Domain', 'SSLfinal_State', 'Domain_registeration_length', 'Favicon', 'port', 'HTTPS_token',]]
```

```
X_train, X_test, y_train, y_test = train_test_split(X_4, y, test_size=0.2, random_state=42)
```

```
skf = StratifiedKFold(n_splits=5)
```

#RandomForest

```
model_random = RandomForestClassifier()
```

```
result_random = cross_validate(model_random, X_train, y_train, scoring={'f1_score' : make_scorer(f1_score),'accuracy_score': make_scorer(accuracy_score),'matthews_corrcoef':make_scorer(matthews_corrcoef)}, cv=skf, n_jobs=-1)
```

```
pd_result_random=pd.DataFrame(result_random)
```

```
pd_result_random
```

```
pd_result_random.mean()
```

#Logistic

```
model_logistic = LogisticRegression()
```

```
result_logistic = cross_validate(model_logistic, X_train, y_train, scoring={'f1_score' : make_scorer(f1_score),'accuracy_score': make_scorer(accuracy_score),'matthews_corrcoef':make_scorer(matthews_corrcoef)}, cv=skf, n_jobs=-1)
```

```
pd_result_logistic = pd.DataFrame(result_logistic)
```

```
pd_result_logistic
```

```
pd_result_logistic.mean()
```

```
model_svm = LinearSVC()
```

```
result_svm = cross_validate(model_svm, X_train, y_train, scoring={'f1_score' : make_scorer(f1_score),'accuracy_score': make_scorer(accuracy_score),'matthews_corrcoef':make_scorer(matthews_corrcoef)}, cv=skf, n_jobs=-1)
```

```
pd_result_svm = pd.DataFrame(result_svm)
```
