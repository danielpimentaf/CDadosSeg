#Pacotes

```
import os
import xml.etree.ElementTree as ET
```

#Permissões por APK

```
path_dir = './manifest/'
list_manifest = os.listdir(path_dir) #Lista de arquivos contidos na pasta manifest
if '.ipynb_checkpoints' in list_manifest:
  list_manifest.remove('.ipynb_checkpoints')
```

```
dict_permission = {} #gerando um dicionário com as permissões por manifest
for manifest in list_manifest:
  path_arquivo = path_dir + manifest
  tree = ET.parse(path_arquivo)
  root = tree.getroot()
  nome_apk, ext = os.path.splitext(manifest)
  nome_apk = nome_apk.replace('AndroidManifest_', '')
  permission = []
  for child in root:
    if child.tag == 'uses-permission':
      string = [*child.attrib.values()]
      for x in string: 
        if x.find('permission') != -1:
          last_string = x
          permission_str = last_string.split('permission')
          permission_str = permission_str[-1].replace('.','',1)
          permission.append(permission_str)
  dict_permission[nome_apk] = permission
```

```
print('############')
print('Permissões por APK')
print('############')
print('')
for key, values in dict_permission.items():
  print('{}: {}'.format(key, values))
  print('')
```

# Permissões Únicas e Comuns

```
list_completa_permission = []
for key, values in dict_permission.items():
  for atributo in values:
    list_completa_permission.append(atributo)
```

```
contagem = {x:list_completa_permission.count(x) for x in list_completa_permission} #Contagem do número de aparições de cada permissão
```

```
permission_unicos = [] #Separação  em únicas e comuns
permission_comum = []
for key, values in contagem.items():
  if values == 1:
    permission_unicos.append(key)  
  else:
    permission_comum.append(key) 
```

```
dict_unicos = {} #Checando qual apk a permissão única pertence
for atributo in permission_unicos:
  for key, values in dict_permission.items():
    if atributo in values:
      if key in dict_unicos:
        dict_unicos[key].append(atributo)
      else:
        dict_unicos[key] = [atributo]
      break
```

```
print('############')
print('Permissões únicas por APK')
print('############')
print('')
for key, values in dict_unicos.items():
  print('{}: {}'.format(key, values))
  print('')
```

```
print('############')
print('Permissões comuns das APKs')
print('############')
print('')
for y in permission_comum:
  print(y)
  print('')
```
