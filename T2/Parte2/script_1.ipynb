```
import pefile
import os
```

```
path_dir = './exe/'
list_exe = os.listdir(path_dir) #Lista de arquivos presentes na pasta exe
if '.ipynb_checkpoints' in list_exe: #remoção de possível arquivo temporário
  list_exe.remove('.ipynb_checkpoints')
```

```
def isSectionExecutable(section): #função para analisar se o arquivo é executável
    characteristics = getattr(section, 'Characteristics')
    if characteristics & 0x00000020 > 0 or characteristics & 0x20000000 > 0:
        return True
    return False

def checkSections(pe):
    list_executaveis = []
    for section in pe.sections:
        if isSectionExecutable(section):
            list_executaveis.append(section.Name.decode().rstrip('\x00'))
    return list_executaveis
```

```
dict_executaveis = {} #Dicionário com as seções executáveis por exe
for exe in list_exe:
  path_arquivo = path_dir + exe
  pe = pefile.PE(path_arquivo)
  section_executaveis=checkSections(pe)
  dict_executaveis[exe] = section_executaveis
```

```
print('############')
print('Seções Executáveis por exe')
print('############')
print('')
for key, values in dict_executaveis.items():
  print('{}: {}'.format(key, values))
  print('')
```
