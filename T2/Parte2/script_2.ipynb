```
import pefile
import os #Pacotes
import sys
```

```
path_dir = './comp/'
list_exe = os.listdir(path_dir) #lista de arquivos na pasta comp
if '.ipynb_checkpoints' in list_exe: #Removendo possíveis arquivos temporários 
  list_exe.remove('.ipynb_checkpoints')
if len(list_exe) != 2: #Checando se há mais de dois arquivos no diretório
  sys.exit("Esperado somente dois arquivos")
```

```
def isSectionExecutable(section): #Seções executáveis
    characteristics = getattr(section, 'Characteristics')
    if characteristics & 0x00000020 > 0 or characteristics & 0x20000000 > 0:
        return True
    return False

def checkSections(pe):
    list_sections = []
    for section in pe.sections:
        if isSectionExecutable(section):
            list_sections.append((section.Name.decode().rstrip('\x00'), True))
        else:
            list_sections.append((section.Name.decode().rstrip('\x00'), False))
    return list_sections
```

```
dict_section = {} #Lista de seções por exe
for exe in list_exe:
  path_arquivo = path_dir + exe
  pe = pefile.PE(path_arquivo)
  list_sections=checkSections(pe)
  dict_section[exe] = list_sections
```

```
a=[*dict_section.values()][0] #Separando as seções por exe
b=[*dict_section.values()][1]
```

```
sections_dif=set(a) ^ set(b) #Seções únicas
sections_iguais = set(a) & set(b) #Seções comuns
```

```
list_a = []
list_b = []
for sections in sections_dif:
  if sections in a:
    list_a.append(sections)
  else:
    list_b.append(sections)
```

```
print('############')
print('Seções Comuns')
print('############')
print('')
for sections in sections_iguais:
  if sections[1]:
    string = 'Executável'
  else:
    string = 'Não Executável'
  print('{} - {}'.format(sections[0],string))
  print('')
print('############')
print('Seções Únicas - {}'.format(list_exe[0]))
print('############')
print('')
for sections in list_a:
  if sections[1]:
    string = 'Executável'
  else:
    string = 'Não Executável'
  print('{} - {}'.format(sections[0],string))
  print('')
print('############')
print('Seções Únicas - {}'.format(list_exe[1]))
print('############')
print('')
for sections in list_b:
  if sections[1]:
    string = 'Executável'
  else:
    string = 'Não Executável'
  print('{} - {}'.format(sections[0],string))
  print('')
```
